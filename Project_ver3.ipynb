```python
# Импорт нужных библиотек - markovify для генерации новых комментариев с помощью марковских цепей, random - для выбора случайного вида спорта
#! pip3 install markovify
import markovify
import random

# Счетчик count - количество вопросов, счетчик points - заработанные баллы
count = 10
points = 0

# Корпуса с комментариями по 4 видам спорта: баскетболу, футболу, волейболу и хоккею.
# Для того, чтобы имена собственные известных людей/команд не выдавали вид спорта, все фамилии были изменены на Иванов/Петров, а все команды - на Спартак/Динамо.

with open("basketball.txt", encoding = "utf-8") as file:
    basketball = file.read()
with open("football.txt", encoding = "utf-8") as file:
    football = file.read()
with open("volleyball.txt", encoding = "utf-8") as file:
    volleyball = file.read()
with open("hockey.txt", encoding = "utf-8") as file:
    hockey = file.read()


#начало программы
print("Нас удивляет, что правила многих популярных видов спорта очень похожи: забей гол, получи очки. \nТакое сходство не могло не отразиться на комментариях к новостям об этих видов спорта. \nМы сгенерировали несколько комментариев на основе реальных комментариев пользователей на популярных спортивных сайтах. \nПопробуйте проверить себя и определить, про какой вид спорта идет речь в каждом из них.")

# Для дальнейшего выбора вида спорта в тесте создан словарь, где ключ - это вид спорта, а значение - соответствующий ему номер правильного ответа
sports = {'футбол':'1','баскетбол' : '2','волейбол' : '3','хоккей':'4'}

#функция для вывода сгенерированных комментариев
def print_gentext():
  print(gentext.make_short_sentence(80), end =' ')
  print(gentext.make_short_sentence(80))


for i in range(1, count+1):
  print('Комментарий №',i)

  # выбор случайного вида спорта

  sport = random.choice(list(sports.keys()))

  # в зависимости от вида спорта, генерация комментариев с помощью ранее объявленной функции
  if sport == 'баскетбол':
    gentext=markovify.Text(basketball, state_size =1, well_formed=True)
    print_gentext()
  elif sport == 'футбол':
    gentext=markovify.Text(football, state_size =1, well_formed=True)
    print_gentext()
  elif sport == 'хоккей':
    gentext=markovify.Text(hockey, state_size =1, well_formed=True)
    print_gentext()
  else:
    gentext=markovify.Text(volleyball, state_size =1, well_formed=True)
    print_gentext()

  print('Какой это вид спорта? Варианты ответа:')
  for key, value in sports.items():
    print(value, '-', key)
  
  # Ввод варианта ответа с клавиатуры
  answer = str(input('Ваш ответ: '))
  if answer == sports.get(sport):
    points +=1
    print('Ура, правильно! Вы получили 1 балл. Ваш счет: ', points, '/10.', sep='')
  else:
    print('Увы, ошибка :( Это был ', sport, '. Ваш счет всё еще ', points, '/10.', sep='')

#конец цикла, вывод итогов игры

print('Конец игры! Ваш результат: ', points, '/10.', sep='', end =' ')
if points < count/3:
  print("Да какая разница, они все просто пинают мяч :)")
elif points < count * 2/3:
  print("Скорее всего, вы сможете отличить мяч от шайбы - уже хорошее достижение!")
elif points < count * 0.9:
  print("Вы отлично разбираетесь в спорте! Можете смело идти комментировать на спортивные сайты.")
else:
  print("Возможно, это вы написали большую часть этих комментариев. Снимаем шляпу!")

```
